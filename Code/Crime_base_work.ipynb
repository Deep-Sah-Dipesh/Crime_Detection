```python
import torch
import torchvision.transforms as transforms
import numpy as np
import cv2
import os
from pytorchvideo.models.hub import i3d_r50

# Load the I3D model pre-trained on the Kinetics dataset
model = i3d_r50(pretrained=True)
model.eval()

def preprocess_video(video_path):
    """
    Preprocess the input video to extract clips of shape (3, 16, 224, 224)
    suitable for I3D input.
    """
    frames = []
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print(f"Error: Cannot open video {video_path}")
        return []

    transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((224, 224)),
        transforms.ToTensor()
    ])

    while True:
        ret, frame = cap.read()
        if not ret:
            break  # No more frames
        if frame is None:
            print("Warning: Skipping empty frame.")
            continue  # Skip empty frames

        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = transform(frame)
        frames.append(frame)

    cap.release()

    if not frames:
        print("Error: No frames were extracted.")
        return []

    frames = torch.stack(frames)  # Shape: (num_frames, 3, 224, 224)

    # Split into clips of 16 frames each
    clips = []
    for i in range(0, len(frames) - 15, 16):  # Sliding window of 16 frames
        clip = frames[i:i+16]
        clips.append(clip)

    if not clips:
        print("Error: No valid clips were extracted.")
        return []

    print(f"Total clips extracted: {len(clips)}")
    return clips  # List of (16, 3, 224, 224) tensors

def extract_video_features(video_path, model, base_save_path=r"C:\Users\rosyd\Crime_Detection\Training\video_features"):
    """
    Extract features from a given video and save them in a folder named
    after the full video name.
    """
    video_name = os.path.splitext(os.path.basename(video_path))[0]  # Preserve full filename
    save_path = os.path.join(base_save_path, video_name)  # Create a folder for each video
    os.makedirs(save_path, exist_ok=True)  # Ensure directory exists

    video_clips = preprocess_video(video_path)  # Preprocess video into 16-frame clips
    
    if not video_clips:
        print(f"Error: No clips generated for video {video_path}")
        return

    features = []

    with torch.no_grad():
        for i, clip in enumerate(video_clips):
            clip = clip.unsqueeze(0)  # Add batch dimension (1, 16, 3, 224, 224)
            clip = clip.permute(0, 2, 1, 3, 4)  # Ensure correct shape (1, 3, 16, 224, 224)

            try:
                feature = model(clip)  # Extract features
                feature = feature.squeeze().detach().cpu().numpy()  # Convert to NumPy
                features.append(feature)

                # Save each extracted feature as a NumPy file
                np.save(os.path.join(save_path, f"feature_{i}.npy"), feature)

            except Exception as e:
                print(f"Error processing clip {i}: {e}")

    print(f"Features saved successfully in: {save_path}")

# Example usage:
video_path = r"C:\Users\rosyd\Crime_Detection\Training\Videos\1-1004\A.Beautiful.Mind.2001__#00-04-20_00-05-35_label_A.mp4"
extract_video_features(video_path, model)
```

```python
import torch
from transformers import RobertaTokenizer, RobertaModel
import os
import numpy as np

# Load the RoBERTa model and tokenizer
def load_roberta_model():
    tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
    model = RobertaModel.from_pretrained('roberta-base')
    return tokenizer, model

# Function to extract text features from the transcript
def extract_text_features(transcript_path, tokenizer, model, base_save_path=r"C:\Users\rosyd\Crime_Detection\Training\text_features"):
    # Load the transcript
    with open(transcript_path, 'r', encoding='utf-8') as file:
        transcript = file.read()

    # Tokenize the transcript and convert to tensor
    inputs = tokenizer(transcript, return_tensors='pt', truncation=True, padding=True, max_length=512)

    # Get the RoBERTa embeddings (last hidden state)
    with torch.no_grad():
        outputs = model(**inputs)

    # The embeddings are in outputs.last_hidden_state
    embeddings = outputs.last_hidden_state  # Shape: (batch_size, seq_len, hidden_size)

    # You can take the [CLS] token (first token) representation as the feature vector
    cls_embedding = embeddings[:, 0, :].cpu().numpy()  # (batch_size, hidden_size)

    # Save the extracted features using the complete transcript name
    transcript_name = os.path.basename(transcript_path)  # Full name with extension
    save_path = os.path.join(base_save_path, transcript_name)  # Save in a folder with the full name

    os.makedirs(save_path, exist_ok=True)  # Ensure the directory exists

    # Save as a NumPy array
    feature_file = os.path.join(save_path, f"{transcript_name}_features.npy")
    np.save(feature_file, cls_embedding)
    print(f"Text features saved successfully in: {feature_file}")

# Load RoBERTa model
tokenizer, model = load_roberta_model()

# Example usage (update with your actual transcript file path)
transcript_path = r"C:\Users\rosyd\Crime_Detection\Training\Transcripts\A.Beautiful.Mind.2001__#00-04-20_00-05-35_label_A.txt"
extract_text_features(transcript_path, tokenizer, model)
```

```python
import os
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader

# Paths to feature folders
video_feature_path = "video_features/"
audio_feature_path = "audio_features/"
text_feature_path = "text_features/"
annotations_path = r"C:\Users\rosyd\Crime_Detection\Training\annotations.txt"

# Load annotations
def load_annotations(annotations_file):
    annotations = {}
    with open(annotations_file, "r") as f:
        for line in f:
            parts = line.strip().split()
            video_name = parts[0]  # First column is video name
            label = int(parts[1])  # Second column is label
            annotations[video_name] = label
    return annotations

annotations = load_annotations(annotations_path)

# Custom Dataset for Meta-Learning
class CrimeDataset(Dataset):
    def __init__(self, video_path, audio_path, text_path, annotations):
        self.video_path = video_path
        self.audio_path = audio_path
        self.text_path = text_path
        self.annotations = annotations
        self.video_list = list(annotations.keys())

    def __len__(self):
        return len(self.video_list)

    def __getitem__(self, idx):
        video_name = self.video_list[idx]

        # Load features
        video_feat = np.load(os.path.join(self.video_path, f"{video_name}.npy"))
        audio_feat = np.load(os.path.join(self.audio_path, f"{video_name}.npy"))
        text_feat = np.load(os.path.join(self.text_path, f"{video_name}.npy"))

        # Concatenate features
        features = np.concatenate([video_feat.flatten(), audio_feat.flatten(), text_feat.flatten()])

        # Get label
        label = self.annotations[video_name]

        return torch.tensor(features, dtype=torch.float32), torch.tensor(label, dtype=torch.long)

# Load dataset
dataset = CrimeDataset(video_feature_path, audio_feature_path, text_feature_path, annotations)
train_loader = DataLoader(dataset, batch_size=16, shuffle=True)
```

```python
import re

# Path to annotations file
annotations_path = r"C:\Users\rosyd\Crime_Detection\Training\annotations.txt"

# Function to load annotations
def load_annotations(annotations_file):
    annotations = {}  # Dictionary to store video -> label mapping
    label_set = set()  # Collect unique labels

    with open(annotations_file, "r") as f:
        for line in f:
            parts = line.strip().split()
            video_name = parts[0]  # First column is video name
            
            # Extract labels from filename using regex
            match = re.search(r"_label_([A-Z0-9\-]+)", video_name)
            if match:
                raw_labels = match.group(1)
                labels = raw_labels.split("-")  # Labels may be multiple (G-B2-B6)
                
                annotations[video_name] = labels
                label_set.update(labels)  # Store all unique labels
    
    return annotations, label_set

# Load annotations
annotations, unique_labels = load_annotations(annotations_path)

# Mapping labels to class indices
label_to_idx = {label: idx for idx, label in enumerate(sorted(unique_labels))}
output_dim = len(label_to_idx)  # Number of crime classes

# Print some results
print(f"Loaded {len(annotations)} videos with labels.")
print(f"Unique crime classes: {unique_labels}")
print(f"Example mapping: {label_to_idx}")
```

```python
import re
from collections import defaultdict

annotations_path = r"C:\Users\rosyd\Crime_Detection\annotations.txt"

# Define a dictionary to store video labels
video_labels = defaultdict(set)  # Use set to avoid duplicate labels

# Read and parse annotations.txt
with open(annotations_path, "r") as file:
    for line in file:
        parts = line.strip().split()  # Split line by spaces
        if len(parts) < 2:
            continue  # Skip empty or malformed lines
        
        video_name = parts[0]  # First part is the video name
        
        # Extract label from the video name using regex
        match = re.search(r"label_([A-Z0-9\-]+)", video_name)
        if match:
            raw_labels = match.group(1).split("-")  # Labels are separated by "-"
            for label in raw_labels:
                if label:  # Avoid empty strings
                    video_labels[video_name].add(label)

# Convert sets to lists for easier processing
video_labels = {video: list(labels) for video, labels in video_labels.items()}

# Print sample results
print(f"Loaded {len(video_labels)} videos with labels.")
print("Example:", list(video_labels.items())[:5])  # Print 5 sample entries
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import os
import random

#  One-Hot Encoding for Labels
def one_hot_encode(labels, label_to_index):
    vector = np.zeros(len(label_to_index))
    for label in labels:
        if label in label_to_index:
            vector[label_to_index[label]] = 1
    return vector

# Define label mapping
label_to_index = {'0': 0, 'B1': 1, 'B2': 2, 'B4': 3, 'B5': 4, 'B6': 5, 'G': 6}
num_classes = len(label_to_index)

#  Load Features for Each Modality
def load_features(video_name):
    video_path = f"video_features/{video_name}.npy"
    audio_path = f"audio_features/{video_name}.npy"
    text_path = f"text_features/{video_name}.npy"
    
    if not os.path.exists(video_path) or not os.path.exists(audio_path) or not os.path.exists(text_path):
        return None

    video_feat = np.load(video_path)
    audio_feat = np.load(audio_path)
    text_feat = np.load(text_path)

    return torch.tensor(video_feat, dtype=torch.float32), torch.tensor(audio_feat, dtype=torch.float32), torch.tensor(text_feat, dtype=torch.float32)

#  Load Labels from File
def load_video_labels(annotation_file="annotations.txt"):
    video_labels = {}
    with open(annotation_file, "r") as f:
        for line in f:
            parts = line.strip().split()
            video_name = parts[0]  # First column is the video name
            labels = parts[1:]  # Remaining columns are labels
            video_labels[video_name] = labels
    return video_labels

#  Dataset Class for Meta-Learning
class CrimeMetaDataset(Dataset):
    def __init__(self, video_labels, k_shot=5, query_size=3):
        self.video_labels = video_labels
        self.video_names = list(video_labels.keys())
        self.k_shot = k_shot
        self.query_size = query_size

    def __len__(self):
        return len(self.video_labels) // (self.k_shot + self.query_size)

    def __getitem__(self, _):
        sampled_videos = random.sample(self.video_names, self.k_shot + self.query_size)
        support_videos = sampled_videos[:self.k_shot]
        query_videos = sampled_videos[self.k_shot:]
        
        support_set, query_set = [], []
        
        for video in support_videos:
            features = load_features(video)
            if features:
                label = one_hot_encode(self.video_labels[video], label_to_index)
                support_set.append((*features, torch.tensor(label, dtype=torch.float32)))
        
        for video in query_videos:
            features = load_features(video)
            if features:
                label = one_hot_encode(self.video_labels[video], label_to_index)
                query_set.append((*features, torch.tensor(label, dtype=torch.float32)))
        
        return support_set, query_set

#  Adaptive Weighting Mechanism
class AdaptiveWeighting(nn.Module):
    def __init__(self, input_dim):
        super(AdaptiveWeighting, self).__init__()
        self.attention = nn.Linear(input_dim, 3)
        self.softmax = nn.Softmax(dim=-1)

    def forward(self, video_feat, audio_feat, text_feat):
        combined = torch.cat([video_feat, audio_feat, text_feat], dim=-1)
        weights = self.softmax(self.attention(combined))
        weighted_features = (weights[:, 0].unsqueeze(1) * video_feat +
                             weights[:, 1].unsqueeze(1) * audio_feat +
                             weights[:, 2].unsqueeze(1) * text_feat)
        return weighted_features

#  Meta-Learner Model
class MetaLearner(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(MetaLearner, self).__init__()
        self.adaptive_weighting = AdaptiveWeighting(input_dim)
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_dim, output_dim)

    def forward(self, video_x, audio_x, text_x):
        fused_features = self.adaptive_weighting(video_x, audio_x, text_x)
        x = self.relu(self.fc1(fused_features))
        return self.fc2(x)

#  MAML Implementation
class MAML:
    def __init__(self, model, lr_inner=0.01, lr_outer=0.001, k_shot=5, meta_batch_size=4):
        self.model = model.to("cuda" if torch.cuda.is_available() else "cpu")
        self.k_shot = k_shot
        self.meta_batch_size = meta_batch_size
        self.inner_lr = lr_inner
        self.outer_lr = lr_outer
        self.optimizer = optim.Adam(self.model.parameters(), lr=self.outer_lr)

    def inner_update(self, support_set):
        criterion = nn.CrossEntropyLoss()
        inner_optimizer = optim.SGD(self.model.parameters(), lr=self.inner_lr)
        loss = None
        has_valid_sample = False
        
        for sample in support_set:
            if len(sample) != 4:
                continue
            
            video_x, audio_x, text_x, y_s = sample
            if video_x is None or audio_x is None or text_x is None:
                continue
            
            has_valid_sample = True
            video_x, audio_x, text_x, y_s = video_x.to("cuda"), audio_x.to("cuda"), text_x.to("cuda"), y_s.to("cuda")
            preds = self.model(video_x, audio_x, text_x)
            loss = criterion(preds, y_s)
            
            inner_optimizer.zero_grad()
            loss.backward()
            inner_optimizer.step()
        
        return loss.item() if has_valid_sample else 0.0
    
    def outer_update(self, query_set):
        meta_loss = torch.tensor(0.0, requires_grad=True, device="cuda" if torch.cuda.is_available() else "cpu")
        criterion = nn.CrossEntropyLoss()

        for video_x, audio_x, text_x, y_q in query_set:
            video_x, audio_x, text_x, y_q = (
                video_x.to("cuda"),
                audio_x.to("cuda"),
                text_x.to("cuda"),
                y_q.to("cuda"),
            )
            preds = self.model(video_x, audio_x, text_x)
            loss = criterion(preds, y_q)
            meta_loss += loss  # Keep accumulating loss as a tensor

        if meta_loss.item() == 0.0:  # No valid samples, avoid calling backward()
            return 0.0

        self.optimizer.zero_grad()
        meta_loss.backward()
        self.optimizer.step()

        return meta_loss.item()


#  Training Loop
def train_maml(video_labels, epochs=5):
    dataset = CrimeMetaDataset(video_labels)
    dataloader = DataLoader(dataset, batch_size=1, shuffle=True)
    
    input_dim = 512 + 1024 + 768
    maml = MAML(MetaLearner(input_dim, hidden_dim=256, output_dim=num_classes))
    
    for epoch in range(epochs):
        for support_set, query_set in dataloader:
            maml.inner_update(support_set)
            maml.outer_update(query_set)
        print(f" Epoch {epoch+1}/{epochs} Complete")
    
    return maml.model

if __name__ == "__main__":
    video_labels = load_video_labels("annotations.txt")
    trained_model = train_maml(video_labels)

```

```python
import os
import whisper

def transcribe_audios_in_directory(base_audio_dir, base_transcript_out_dir, model_name="base"):
    print(f"\nLoading Whisper model: {model_name}...")
    try:
        model = whisper.load_model(model_name)
        print("Whisper model loaded successfully.")
    except Exception as e:
        print(f" Error loading Whisper model: {e}")
        return

    print(f"Starting transcription for audio files in: {base_audio_dir}")
    print(f"Transcripts will be saved to: {base_transcript_out_dir}")

    audio_extensions = ('.mp3', '.wav', '.m4a')
    
    print("Counting total audio files to process...")
    total_audios = count_files_by_extension(base_audio_dir, audio_extensions)
    if total_audios == 0:
        print("No audio files found to process.")
        return
    print(f"Found {total_audios} audio file(s).")

    newly_processed_count = 0
    skipped_count = 0
    failed_count = 0
    current_audio_num = 0

    for dirname, _, filenames in os.walk(base_audio_dir):
        for filename in filenames:
            if filename.lower().endswith(audio_extensions):
                current_audio_num += 1
                audio_path = os.path.join(dirname, filename)
                
                relative_path = os.path.relpath(dirname, base_audio_dir)
                transcript_output_subdir = os.path.join(base_transcript_out_dir, relative_path)
                os.makedirs(transcript_output_subdir, exist_ok=True)
                
                base_filename_no_ext = os.path.splitext(filename)[0]
                transcript_filename = f"{base_filename_no_ext}.txt"
                transcript_path = os.path.join(transcript_output_subdir, transcript_filename)

                print(f"\n[{current_audio_num}/{total_audios}] Checking audio: {filename}")

                if os.path.exists(transcript_path):
                    print(f" SKIPPED: Transcript file already exists at {transcript_path}")
                    skipped_count += 1
                    continue

                print(f"Input: {audio_path} -> Output: {transcript_path}")
                try:
                    result = model.transcribe(audio_path, fp16=False) 
                    with open(transcript_path, "w", encoding="utf-8") as file:
                        file.write(result["text"])
                    print(f" Transcript saved: {transcript_path}")
                    newly_processed_count += 1
                except Exception as e:
                    print(f" Error transcribing {filename}: {e}")
                    failed_count += 1
                
    print(f"\n--- Transcription Summary ---")
    print(f"Total audio files checked: {total_audios}")
    print(f"Newly transcribed: {newly_processed_count}")
    print(f"Skipped (already existed): {skipped_count}")
    print(f"Failed to transcribe: {failed_count}")

# Call the function
transcribe_audios_in_directory(BASE_AUDIO_OUTPUT_DIR, BASE_TRANSCRIPT_OUTPUT_DIR)
```

```python
import os
import subprocess

def count_files_by_extension(directory, extensions):
    count = 0
    for _, _, filenames in os.walk(directory):
        for filename in filenames:
            if filename.lower().endswith(extensions):
                count += 1
    return count

def extract_audio_from_videos(base_video_dir, base_audio_out_dir):
    print(f"Starting audio extraction from: {base_video_dir}")
    print(f"Audio will be saved to: {base_audio_out_dir}")
    
    video_extensions = ('.mp4', '.avi', '.mov', '.mkv')
    
    print("Counting total videos to process...")
    total_videos = count_files_by_extension(base_video_dir, video_extensions)
    if total_videos == 0:
        print("No video files found to process.")
        return
    print(f"Found {total_videos} video file(s).")

    processed_count = 0
    failed_count = 0
    current_video_num = 0

    for dirname, _, filenames in os.walk(base_video_dir):
        for filename in filenames:
            if filename.lower().endswith(video_extensions):
                current_video_num += 1
                video_path = os.path.join(dirname, filename)
                
                relative_path = os.path.relpath(dirname, base_video_dir)
                audio_output_subdir = os.path.join(base_audio_out_dir, relative_path)
                os.makedirs(audio_output_subdir, exist_ok=True)
                
                base_filename_no_ext = os.path.splitext(filename)[0]
                audio_filename = f"{base_filename_no_ext}.mp3"
                audio_path = os.path.join(audio_output_subdir, audio_filename)

                print(f"\nProcessing video {current_video_num} of {total_videos}: {filename}")
                print(f"Input: {video_path}")
                
                ffmpeg_command = [
                    'ffmpeg', '-i', video_path, 
                    '-q:a', '0', '-map', 'a', 
                    audio_path, '-y'
                ]
                
                try:
                    process = subprocess.run(ffmpeg_command, check=True, capture_output=True, text=True)
                    if os.path.exists(audio_path):
                        print(f" Audio extraction successful: {audio_path}")
                        processed_count += 1
                    else:
                        print(f" Audio extraction reported success, but file not found: {audio_path}")
                        failed_count +=1
                except subprocess.CalledProcessError as e:
                    print(f" Error extracting audio from {filename}:")
                    # print(f"Command: {' '.join(e.cmd)}") # Can be verbose
                    print(f"Stderr: {e.stderr.strip()}")
                    failed_count += 1
                except Exception as e:
                    print(f" An unexpected error occurred with {filename}: {e}")
                    failed_count += 1
                # print("-" * 30) # Use newline from start of print for separation

    print(f"\n--- Audio Extraction Summary ---")
    print(f"Successfully processed: {processed_count} of {total_videos} videos.")
    print(f"Failed to process: {failed_count} videos.")

# Call the function
extract_audio_from_videos(BASE_VIDEO_INPUT_DIR, BASE_AUDIO_OUTPUT_DIR)
```

```python
import os
import numpy as np
import librosa
import tensorflow as tf
import tensorflow_hub as hub
import pandas as pd
from tqdm import tqdm
import warnings
import shutil
import tempfile

class RobustAudioFeatureExtractor:
    """
    Enhanced YAMNet-based audio feature extractor with multiple fallback strategies
    """
    
    def __init__(self, cache_dir=None):
        self.sample_rate = 16000
        self.model = None
        self.class_names = None
        self.cache_dir = cache_dir or r'G:\Crime_Detection\Training\tfhub_cache'
        
        # Set persistent cache directory
        os.environ['TFHUB_CACHE_DIR'] = self.cache_dir
        os.makedirs(self.cache_dir, exist_ok=True)
        
        # Load model with fallback strategies
        self._load_model_with_fallback()
    
    def _load_model_with_fallback(self):
        """Load YAMNet with multiple fallback strategies"""
        strategies = [
            self._load_from_hub,
            self._load_after_cache_clear,
            self._load_with_retry
        ]
        
        for i, strategy in enumerate(strategies, 1):
            try:
                print(f"Attempting loading strategy {i}...")
                if strategy():
                    print(f" Successfully loaded YAMNet using strategy {i}")
                    return
            except Exception as e:
                print(f" Strategy {i} failed: {e}")
                continue
        
        raise RuntimeError("All YAMNet loading strategies failed")
    
    def _load_from_hub(self):
        """Strategy 1: Load directly from TensorFlow Hub"""
        self.model = hub.load('https://tfhub.dev/google/yamnet/1')
        self._load_class_names()
        return True
    
    def _load_after_cache_clear(self):
        """Strategy 2: Clear cache and reload"""
        # Clear corrupted cache directories
        default_cache = os.path.join(tempfile.gettempdir(), 'tfhub_modules')
        if os.path.exists(default_cache):
            shutil.rmtree(default_cache)
        
        if os.path.exists(self.cache_dir):
            shutil.rmtree(self.cache_dir)
            os.makedirs(self.cache_dir, exist_ok=True)
        
        # Reload model
        self.model = hub.load('https://tfhub.dev/google/yamnet/1')
        self._load_class_names()
        return True
    
    def _load_with_retry(self):
        """Strategy 3: Multiple retry attempts"""
        for attempt in range(3):
            try:
                print(f"  Retry attempt {attempt + 1}/3")
                self.model = hub.load('https://tfhub.dev/google/yamnet/1')
                self._load_class_names()
                return True
            except Exception as e:
                if attempt == 2:  # Last attempt
                    raise e
                print(f"  Attempt {attempt + 1} failed, retrying...")
                continue
        return False
    
    def _load_class_names(self):
        """Load YAMNet class names with error handling"""
        try:
            class_map_path = self.model.class_map_path().numpy().decode('utf-8')
            class_map_csv = pd.read_csv(class_map_path)
            self.class_names = class_map_csv['display_name'].tolist()
            print(f" Loaded {len(self.class_names)} class names")
        except Exception as e:
            print(f"Warning: Could not load class names: {e}")
            # Create dummy class names as fallback
            self.class_names = [f"class_{i}" for i in range(521)]
    
    def preprocess_audio(self, audio_path):
        """Preprocess audio file for YAMNet"""
        try:
            waveform, sr = librosa.load(audio_path, sr=self.sample_rate, mono=True)
            waveform = waveform.astype(np.float32)
            return waveform
        except Exception as e:
            print(f" Error preprocessing audio {audio_path}: {e}")
            return None
    
    def extract_features(self, waveform):
        """Extract YAMNet features from preprocessed waveform"""
        try:
            scores, embeddings, spectrogram = self.model(waveform)
            
            return {
                'embeddings': embeddings.numpy(),
                'scores': scores.numpy(),
                'spectrogram': spectrogram.numpy(),
                'class_names': self.class_names
            }
        except Exception as e:
            print(f" Error extracting features: {e}")
            return None
    
    def get_top_classes(self, scores, top_k=5):
        """Get top-k predicted classes for each audio segment"""
        top_classes_info = []
        
        for i, segment_scores in enumerate(scores):
            top_indices = np.argsort(segment_scores)[-top_k:][::-1]
            
            segment_info = {
                'segment_idx': i,
                'time_start': i * 0.48,  # YAMNet frame rate
                'top_classes': []
            }
            
            for rank, class_idx in enumerate(top_indices):
                class_info = {
                    'rank': rank + 1,
                    'class_name': self.class_names[class_idx],
                    'score': float(segment_scores[class_idx])
                }
                segment_info['top_classes'].append(class_info)
            
            top_classes_info.append(segment_info)
        
        return top_classes_info
    
    def save_features(self, features, output_dir, filename_base):
        """Save extracted features to files"""
        os.makedirs(output_dir, exist_ok=True)
        
        # Save embeddings
        embeddings_path = os.path.join(output_dir, f"{filename_base}_yamnet_embeddings.npy")
        np.save(embeddings_path, features['embeddings'])
        
        # Save class predictions
        top_classes = self.get_top_classes(features['scores'])
        csv_data = []
        
        for segment in top_classes:
            for class_info in segment['top_classes']:
                csv_data.append({
                    'segment_index': segment['segment_idx'],
                    'time_start_sec': segment['time_start'],
                    'rank': class_info['rank'],
                    'class_name': class_info['class_name'],
                    'score': class_info['score']
                })
        
        if csv_data:
            df = pd.DataFrame(csv_data)
            csv_path = os.path.join(output_dir, f"{filename_base}_yamnet_classes.csv")
            df.to_csv(csv_path, index=False)
        
        # Save spectrogram
        spectrogram_path = os.path.join(output_dir, f"{filename_base}_yamnet_spectrogram.npy")
        np.save(spectrogram_path, features['spectrogram'])
        
        return {
            'embeddings_path': embeddings_path,
            'csv_path': csv_path if csv_data else None,
            'spectrogram_path': spectrogram_path
        }

# Updated main extraction function
def extract_audio_features_from_directory(base_audio_dir, base_features_output_dir, 
                                        audio_extensions=('.mp3', '.wav', '.m4a')):
    """
    Extract audio features using the robust extractor
    """
    print("="*80)
    print("STARTING ROBUST AUDIO FEATURE EXTRACTION")
    print("="*80)
    
    # Initialize robust feature extractor
    extractor = RobustAudioFeatureExtractor()
    
    # Count total files
    total_files = 0
    for root, dirs, files in os.walk(base_audio_dir):
        for file in files:
            if file.lower().endswith(audio_extensions):
                total_files += 1
    
    print(f"Found {total_files} audio files to process")
    
    if total_files == 0:
        print("No audio files found!")
        return
    
    processed_count = 0
    skipped_count = 0
    failed_count = 0
    
    with tqdm(total=total_files, desc="Processing audio files") as pbar:
        for root, dirs, files in os.walk(base_audio_dir):
            for filename in files:
                if filename.lower().endswith(audio_extensions):
                    
                    audio_path = os.path.join(root, filename)
                    relative_dir = os.path.relpath(root, base_audio_dir)
                    
                    filename_base = os.path.splitext(filename)[0]
                    output_dir = os.path.join(base_features_output_dir, relative_dir, filename_base)
                    
                    # Check if already processed
                    embeddings_path = os.path.join(output_dir, f"{filename_base}_yamnet_embeddings.npy")
                    if os.path.exists(embeddings_path):
                        pbar.set_description(f"Skipped: {filename}")
                        skipped_count += 1
                        pbar.update(1)
                        continue
                    
                    pbar.set_description(f"Processing: {filename}")
                    
                    try:
                        # Preprocess audio
                        waveform = extractor.preprocess_audio(audio_path)
                        if waveform is None:
                            raise ValueError("Failed to preprocess audio")
                        
                        # Extract features
                        features = extractor.extract_features(waveform)
                        if features is None:
                            raise ValueError("Failed to extract features")
                        
                        # Save features
                        saved_paths = extractor.save_features(features, output_dir, filename_base)
                        processed_count += 1
                        
                    except Exception as e:
                        print(f"\n Error processing {filename}: {e}")
                        failed_count += 1
                    
                    pbar.update(1)
    
    print("\n" + "="*80)
    print("ROBUST AUDIO FEATURE EXTRACTION COMPLETED")
    print(f"Total files: {total_files}")
    print(f"Successfully processed: {processed_count}")
    print(f"Skipped (already existed): {skipped_count}")
    print(f"Failed: {failed_count}")
    print("="*80)

# Main execution with updated paths
if __name__ == "__main__":
    BASE_AUDIO_DIR = r"G:\Crime_Detection\Training\Processed_Data\Audios"
    BASE_FEATURES_OUTPUT_DIR = r"G:\Crime_Detection\Training\Processed_Data\audio_features"
    
    os.makedirs(BASE_FEATURES_OUTPUT_DIR, exist_ok=True)
    
    print(f"Input audio directory: {BASE_AUDIO_DIR}")
    print(f"Output features directory: {BASE_FEATURES_OUTPUT_DIR}")
    
    if not os.path.exists(BASE_AUDIO_DIR):
        print(f" Error: Input directory does not exist: {BASE_AUDIO_DIR}")
    else:
        extract_audio_features_from_directory(BASE_AUDIO_DIR, BASE_FEATURES_OUTPUT_DIR)
```
